```python
def to_str(x):
    try:
        a = str(int(x))
    except:
        a = x
    return a

df['BUILDING_NUMBER']=df['BUILDING_NUMBER'].apply(lambda x:to_str(x))
df['SUB_BUILDING_NAME']=df['SUB_BUILDING_NAME'].apply(lambda x:to_str(x))
df['BUILDING_NAME']=df['BUILDING_NAME'].apply(lambda x:to_str(x))
```

```python
price_all_base = matching.change_df(price_all_base,3)
```

```python
def find_one(df,price_all_base,x):
    filt_postcode = matching.find_by_postcode(price_all_base,df.iloc[x]['POSTCODE_LOCATOR'])
    
    for i in range(len(filt_postcode.index)):
        
        if pd.isnull(filt_postcode.iloc[i][7]):
            if pd.isnull(filt_postcode.iloc[i][8]):
                continue
            if str(filt_postcode.iloc[i][8]) == str(df.iloc[x]['SUB_BUILDING_NAME']):
                return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]]
            continue
        name_num = filt_postcode.iloc[i][7]
        name_num_lst = name_num.split(', ')
        if len(name_num_lst) == 1:
            cur = name_num_lst[0]
            if cur[0].isdigit():
                if cur == df.iloc[x]['BUILDING_NUMBER']:
                    return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]]
                continue
            else:
                if cur == df.iloc[x]['BUILDING_NAME']:
                    if pd.isnull(df.iloc[x]['SUB_BUILDING_NAME']):
                        return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]]
                    elif df.iloc[x]['SUB_BUILDING_NAME'] == filt_postcode.iloc[i][8]:
                        return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]]
        else:
            if name_num_lst[0] == df.iloc[x]['BUILDING_NAME']:
                return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]]
            if name_num_lst[0] == df.iloc[x]['BUILDING_NUMBER']:
                return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]]
            if name_num_lst[1] == df.iloc[x]['BUILDING_NAME']:
                return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]]
            if name_num_lst[1] == df.iloc[x]['BUILDING_NUMBER']:
                return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]]
    
    for i in range(len(filt_postcode.index)):
        name_num = filt_postcode.iloc[i][7]
        name_num_lst = str(name_num).split()
        #print(name_num_lst)
        name_num_lst.append(filt_postcode.iloc[i][8])
        s1 = set(name_num_lst)
        a = str(df.iloc[x]['BUILDING_NUMBER']).split()
        b = str(df.iloc[x]['BUILDING_NAME']).split()
        c = str(df.iloc[x]['SUB_BUILDING_NAME']).split()
        a.extend(b)
        a.extend(c)
        #print(a)
        s2 = set(list(a))
        if np.nan in s1:
            s1.remove(np.nan)
        if np.nan in s2:
            s2.remove(np.nan)
        if 'nan' in s1:
            s1.remove('nan')
        if 'nan' in s2:
            s2.remove('nan')
        #print(s1,s2)
        if s1 == s2 and len(s1)!=0:
            return [filt_postcode.iloc[i][1],filt_postcode.iloc[i][2]] 
    return ['Not Found','Not Found']
```

```python
find_one(df,price_all_base,35)
```

```python
df['last_sold_price'] = '-'
df['last_sold_date'] = '-'
```

```python
for x in range(len(df.index)):
    res = find_one(df,price_all_base,x)
    df.loc[x,'last_sold_price'] = res[0]
    df.loc[x,'last_sold_date'] = res[1]
    if x%1000 == 0:
        print(x)
    if x%100000 == 0:
        df.to_excel('result.xlsx')
    elif x%10000 == 0:
        df.to_csv('result.csv')
```
